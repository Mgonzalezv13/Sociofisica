Explicación del modelo
--


El modelo del umbral (o de Granovetter) consiste en simular la dinamica de una manifestacion y como en funcion del número de personas que participan de esta, se suman más personas.

Cada individuo tiene un umbral propio, en el cual evaluara en funcion del número de personas que están participando de la manifestacion, si es que este se suma o se mantiene al margen.

De esta forma, podemos caracterizar la cantidad de personas que se irán sumando a la manifestacion en un tiempo $t+1$ a partir de la cantidad de personas que forman parte de esta en un tiempo $t$, a través de la siguiente relación de recurrencia.

$r(t+1) = F[r(t)]$

Dónde $r(t)$ corresponde a la cantidad de personas que se sumaron a la manifestación en un tiempo $t$, $F[r(t)]$ corresponde a la suma total de personas que se sumaron en los tiempos 0,1,2... hasta llegar a $t$ y $r(t+1)$ a la cantidad de personas que se sumarán en un tiempo $t+1$

Explicacion de la simulación
--

1. En la primera celda se define la función que corre la simulación, esta recibe 3 parametros los cuales son una distribucion, el número total de agentes y el número de umbrales. Los últimos dos parametros tienen valores prefijados, pero se pueden cambiar si es que se entrega más de un parametro al momento de llamar a la funcion modelo_umbral
2. En las siguientes dos celdas, se define una distribucion y se llama a la función entregando dicha distribucion como parametro.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def modelo_umbral(distribucion,N=1000,umbrales=100): #distribucion corresponde al tipo de dist que se quiere estudiar
    umb = np.linspace(0,100,umbrales)                #N es el numero total de individuos
    manifestantes = []                                          
    for i in range(umbrales):                           
      amarillos = 0
      movilizados = 0
      for j in range(N):
        if distribucion[j]>i:#se chequea la condicion para unirse a la manifestacion en funcion de cada individuo
          amarillos += 1
        else:
          movilizados += 1
      manifestantes.append(movilizados/10)

    plt.plot(umb,umb)
    plt.plot(umb, manifestantes, color='red')
    plt.title('Modelo del umbral')
```

```python
#Ploteo para el caso de una distribucion normal
N=1000
dis_normal = np.abs(np.random.normal(0.4,0.2,N))*100
modelo_umbral(dis_normal)
```

```python
#Ploteo para el caso de una distribucion uniforme
dis_uniforme = np.random.uniform(0,100,N)
modelo_umbral(dis_uniforme)
```
